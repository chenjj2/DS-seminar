```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_s = 1000.; f = 1.
x = np.arange(0.,20.,1/f_s)

gauss = np.random.normal(0., 1., size=len(x))
plt.plot(x, gauss, 'b.'); #plt.xlim([0,2])
```

```python
fft = np.fft.fft(gauss)  
fft_shifted = np.fft.fftshift(fft)  
freq = np.fft.fftfreq(len(fft), 1/f_s)
freq_shifted = np.fft.fftshift(freq)
plt.plot(freq_shifted, np.abs(fft_shifted)**10.)
plt.xlim([0,20]); plt.ylim([0,1e23])
plt.plot([2,2,4,4,6,6],[0,1e23,1e23,0,0,1e23])
```

```python
scale = np.sin(2*2*np.pi*f*x)
gauss_scale = gauss*scale
plt.plot(x, gauss_scale, 'b.'); plt.plot(x, scale, 'r'); plt.xlim([0,2])
```

```python
fft = np.fft.fft(gauss_scale)
freq = np.fft.fftfreq(len(fft), 1/f_s)
fft_shifted = np.fft.fftshift(fft)
freq_shifted = np.fft.fftshift(freq)
plt.plot(freq_shifted, np.abs(fft_shifted)**10.); 
plt.xlim([0,20]); plt.ylim([0,1e23])
plt.plot([2,2,4,4,6,6],[0,1e23,1e23,0,0,1e23])
```
