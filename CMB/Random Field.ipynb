```python
%matplotlib inline
import pyfits
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
f = pyfits.open('camb_28841594_scalcls.fits')
power = f[1].data.field('temperature')
n_modes = np.size(power)
# flatten the first 10 modes 
yamp = power**0.5; yamp[0:10] = yamp[0:10]*0.0
# sample spectra at *10 the number of modes
samp = 10.0; n = n_modes * samp; xax = np.arange(n)/samp
```

```python
for i in range(10,20,4):
    phase = np.random.uniform(0, np.pi*2)
    a_tmp = yamp[i]*np.sin(xax*np.pi*i/n_modes+phase)
    plt.plot(a_tmp, label=r'%3.1f $\pi$' %(phase/np.pi))
plt.title('example of 3 modes with random phase')
plt.ylabel('Temperatrue')
plt.legend()
```

```python
temp = np.zeros(n)
for freq in range(n_modes):
    phase = np.random.uniform(0, np.pi*2)
    a_tmp = yamp[freq]*np.sin(xax*np.pi*freq/n_modes+phase)
    temp = temp+a_tmp
plt.plot(temp); plt.ylabel('Temperature')
```

```python
plt.hist(temp); plt.title('Temperature Distribution')
```

```python

fft = np.abs(np.fft.fftshift(np.fft.fft(temp))); fft_len = len(fft)
fft_plot = np.log10(fft[int(fft_len/2):fft_len]*10**-4) 
plt.plot(np.arange(1000)*2,(fft_plot[0:1000])**1.*(np.arange(1000)*2)**2,'.'); plt.xlim([10,2000])
plt.plot(np.log10(yamp), linewidth=2., color='red');
```
