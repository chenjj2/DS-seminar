Fourier Transform: time series h(t), freq series H(f)
$$ H(f)=\int^\infty_{-\infty}h(t)exp(-i2\pi ft)dt$$
Inverse Fourier Transform
$$ h(t)=\int^\infty_{-\infty}H(f)exp(i2\pi ft)df$$
One-sided PSD(power spectral density)
$$ PSD(f) := |H(f)|^2+|H(-f)|^2 $$
Convolution
$$ (a \ast b)(t)=\int^\infty_{-\infty} a(t^{'})b(t-t^{'})dt^{'}$$
Convolution Theorem
$$ h=a\ast b$$
$$ H(f)=A(f)B(f)$$

Nyquist sampling theorem
$$ f_c = 2 max(f)$$

```python
#f_s = 45.0 # Hz
f_s = 35.0
f1 = 1.0; f2 = 20. # Hz
time = np.arange(0.0, 3.0, 1/f_s)
x = 5 * np.sin(2 * np.pi * f1 * time) + 2 * np.sin(2 * np.pi * f2 * time)
plt.plot(time, x)
plt.xlabel("Time (sec)")
plt.ylabel("x")
```

```python
fft_x = np.fft.fft(x)
n = len(fft_x)
freq = np.fft.fftfreq(n, 1/f_s)
fft_x_shifted = np.fft.fftshift(fft_x)
freq_shifted = np.fft.fftshift(freq)
```

```python
plt.plot(freq_shifted, np.abs(fft_x_shifted))
plt.xlabel("Frequency (Hz)")
```
