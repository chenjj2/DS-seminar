```python
Nt = 4
```

```python
tmax=max(t); tmin=min(t)
tcut = np.linspace(tmin,tmax,Nt+1)
```

```python
t_table = np.empty((Nt,np.size(t)),dtype=bool)
c_table = np.zeros(Nt)
```

```python
for i in range(Nt):
    t_table[i,:] = (t<tcut[i+1]) & (t>tcut[i])
    c_table[i] = np.sum(t_table[i,:])
```

```python
f,ax = plt.subplots(1,4,figsize=(19,4))
(ax1,ax2,ax3,ax4) = ax
for i in range(Nt):
    ax[i].plot(ra[t_table[i,:]],dec[t_table[i,:]],'b.',alpha=0.01); ax[i].axis('equal')
    ax[i].set_xlim([xmin,xmax]); ax[i].set_ylim([ymin,ymax])
```

```python
plt.plot(c_table,'x-')
plt.title('Counts in each Time Bin')
```

```python
Nx = 100
Ny = 100
Nneighbor = 100

#------------------------------------------------------------
# Evaluate for several models
Xgrid = np.vstack(map(np.ravel, np.meshgrid(np.linspace(xmin, xmax, Nx),
                                            np.linspace(ymin, ymax, Ny)))).T
```

```python
Nimage = np.zeros((Nt,Nx,Ny))
```

```python
knn = KNeighborsDensity('bayesian', Nneighbor)
```

```python
for i in range(Nt):
    ra1,dec1 = ra[t_table[i,:]], dec[t_table[i,:]]
    X = np.vstack((ra1,dec1)).T
    dens = knn.fit(X).eval(Xgrid).reshape(Ny,Nx)
    Nimage[i,:,:]=dens
```

```python
f,ax = plt.subplots(1,4,figsize=(19,4))
(ax1,ax2,ax3,ax4) = ax

for i in range(Nt):
    ax[i].imshow(Nimage[i,:,:],norm=LogNorm())
```
