This project tries to find out how to identify time-varying sources with photon data. 

For imaging data, we can subtract the images of the source taken at two different periods and identify the residual of image subtraction. But it does not make as much sense to 'subtract two images' for photon data because photon data usually have great spatial precision that there are barely any two photons that has the same position. We can change this situation by smoothing the photon images. This project experiments with three methods(2D histogram, nearest neighbor, kernel density estimation) that smooths the photon images and compare how they perform.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
%matplotlib inline
from astroML.density_estimation import KNeighborsDensity
from sklearn.neighbors import KernelDensity
from matplotlib.colors import LogNorm
import pyfits
```

```python
t1,a,b,c,d,e,f,g,ra1,dec1,h = np.loadtxt('J011824.47+344505.0/e13484-nd_30.0as.csv',delimiter=',',unpack=True)
t2,a,b,c,d,e,f,g,ra2,dec2,h = np.loadtxt('J011824.47+344505.0/e18102-nd_30.0as.csv',delimiter=',',unpack=True)
```

```python
tmax1 = max(t1); tmin1 = min(t1)
xmax1, xmin1, ymax1, ymin1 = max(ra1), min(ra1), max(dec1), min(dec1)
tmax2 = max(t2); tmin2 = min(t2)
xmax2, xmin2, ymax2, ymin2 = max(ra2), min(ra2), max(dec2), min(dec2)
```

```python
f,(ax1,ax2,ax3,ax4)=plt.subplots(1,4,figsize=(19,4))

ax1.plot(ra1,dec1,'b.',alpha=0.1,rasterized=True)
ax1.axis('equal'); ax1.set_xlim([xmin1,xmax1]); ax1.set_ylim([ymin1,ymax1])
ax1.set_title('All Photons [Period A]')

ax2.hist(t1, bins=np.linspace(tmin1,tmax1,5))
ax2.set_title('Counts/Time Bin [A]')

ax3.plot(ra2,dec2,'b.',alpha=0.1,rasterized=True)
ax3.axis('equal'); ax3.set_xlim([xmin2,xmax2]); ax3.set_ylim([ymin2,ymax2])
ax3.set_title('All Photons [Period B]')

ax4.hist(t2, bins=np.linspace(tmin2,tmax2,5))
ax4.set_title('Counts/Time Bin [B]')
```

Pick the same point source at two different periods. 

During period A, the source's brightness is changing; while during period B, the source remains constant.

```python
Nt = 4
```

```python
tcut1 = np.linspace(tmin1,tmax1,Nt+1)
tcut2 = np.linspace(tmin2,tmax2,Nt+1)
```

```python
t_table1 = np.empty((Nt,np.size(t1)),dtype=bool)
t_table2 = np.empty((Nt,np.size(t2)),dtype=bool)
```

```python
for i in range(Nt):
    t_table1[i,:] = (t1<tcut1[i+1]) & (t1>tcut1[i])
    t_table2[i,:] = (t2<tcut2[i+1]) & (t2>tcut2[i])
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax
for i in range(Nt):
    ax[0][i].plot(ra1[t_table1[i,:]],dec1[t_table1[i,:]],'b.',alpha=0.2); ax[0][i].axis('equal')
    ax[0][i].set_xlim([xmin1,xmax1]); ax[0][i].set_ylim([ymin1,ymax1])
    ax[0][i].set_title('Bin '+str(i))
    
    ax[1][i].plot(ra2[t_table2[i,:]],dec2[t_table2[i,:]],'b.',alpha=0.2); ax[1][i].axis('equal')
    ax[1][i].set_xlim([xmin2,xmax2]); ax[1][i].set_ylim([ymin2,ymax2])

ax[0][0].set_ylabel('A')
ax[1][0].set_ylabel('B')
```

The photons are split into four time bins. The top 4 are photon plots during period A; the bottom are photon plots during period B.

=====================================

2D histogram

```python
Nx = 30
Ny = 30

xedges1 = np.linspace(xmin1,xmax1,Nx+1)
yedges1 = np.linspace(ymin1,ymax1,Ny+1)

xedges2 = np.linspace(xmin2,xmax2,Nx+1)
yedges2 = np.linspace(ymin2,ymax2,Ny+1)
```

```python
Nimage1 = np.zeros((Nt,Nx,Ny))
Nimage2 = np.zeros((Nt,Nx,Ny))
```

```python
for i in range(Nt):
    ra11,dec11 = ra1[t_table1[i,:]], dec1[t_table1[i,:]]
    H, xedges, yedges = np.histogram2d(ra11, dec11, bins=(xedges1, yedges1))
    Nimage1[i,:,:] = H
    
    ra21,dec21 = ra2[t_table2[i,:]], dec2[t_table2[i,:]]
    H, xedges, yedges = np.histogram2d(ra21, dec21, bins=(xedges2, yedges2))
    Nimage2[i,:,:] = H
```

```python
Nimagemax, Nimagemin = max(np.max(Nimage1),np.max(Nimage2)), min(np.min(Nimage1),np.min(Nimage2))
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax

for i in range(Nt):
    im1=ax[0][i].imshow(Nimage1[i,:,:],origin='lower',vmin=Nimagemin,vmax=Nimagemax)
    ax[0][i].set_title('Bin '+str(i))
    im2=ax[1][i].imshow(Nimage2[i,:,:],origin='lower',vmin=Nimagemin,vmax=Nimagemax)

ax[0][0].set_ylabel('A')
ax[1][0].set_ylabel('B')

cbar_ax2 = f.add_axes([0.95, 0.15, 0.02, 0.7])
f.colorbar(im2,cax=cbar_ax2)
```

Smoothing photon images with 2D histogram.

```python
Ndiff1 = np.zeros_like(Nimage1)
Ndiff2 = np.zeros_like(Nimage2)
for i in range(Nt):
    Ndiff1[i,:,:] = Nimage1[i,:,:]-Nimage1[0,:,:]
    Ndiff2[i,:,:] = Nimage2[i,:,:]-Nimage2[0,:,:]
```

```python
Ndiffmax, Ndiffmin = max(np.max(Ndiff1),np.max(Ndiff2)), min(np.min(Ndiff1),np.min(Ndiff2))
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax

for i in range(Nt):
    im1=ax[0][i].imshow(Ndiff1[i,:,:], origin='lower',vmin=Ndiffmin,vmax=Ndiffmax)
    ax[0][i].set_title('Bin '+str(i))
    im2=ax[1][i].imshow(Ndiff2[i,:,:], origin='lower',vmin=Ndiffmin,vmax=Ndiffmax)
    
ax[0][0].set_ylabel('A')
ax[1][0].set_ylabel('B')
    
cbar_ax2 = f.add_axes([0.95, 0.15, 0.02, 0.7])
f.colorbar(im2,cax=cbar_ax2)
```

Residual after subtracting the image of Bin 0.

```python
ratio1 = np.zeros(3)
ratio2 = np.zeros(3)
for i in range(1,Nt):
    ratio1[i-1] = np.max(np.abs(Ndiff1[i,:,:]))/np.max(np.abs(Nimage1[i,:,:]))
    ratio2[i-1] = np.max(np.abs(Ndiff2[i,:,:]))/np.max(np.abs(Nimage2[i,:,:]))
```

```python
a,=plt.plot(range(1,4),ratio1,'r+')
b,=plt.plot(range(1,4),ratio2,'gx')
plt.xlim([0.5,3.5]);plt.xlabel('Bin');plt.ylabel('Residual / Full amplitude')
plt.legend((a,b),('A','B'),loc=2)
plt.yscale('log')
```

```python
print ratio1, ratio2
```

[0.31111111, 5.11111111]

=====================================

Nearest Neighbor

```python
Nneighbor = 100
```

```python
Nx = 100
Ny = 100

Xgrid1 = np.vstack(map(np.ravel, np.meshgrid(np.linspace(xmin1, xmax1, Nx), np.linspace(ymin1, ymax1, Ny)))).T
Xgrid2 = np.vstack(map(np.ravel, np.meshgrid(np.linspace(xmin2, xmax2, Nx), np.linspace(ymin2, ymax2, Ny)))).T

knn = KNeighborsDensity('bayesian', Nneighbor)
```

```python
Nimage1 = np.zeros((Nt,Nx,Ny))
Nimage2 = np.zeros((Nt,Nx,Ny))
```

```python
for i in range(Nt):
    ra11,dec11 = ra1[t_table1[i,:]], dec1[t_table1[i,:]]
    X1 = np.vstack((ra11,dec11)).T
    dens1 = knn.fit(X1).eval(Xgrid1).reshape(Ny,Nx)
    Nimage1[i,:,:]=dens1
    
    ra21,dec21 = ra2[t_table2[i,:]], dec2[t_table2[i,:]]
    X2 = np.vstack((ra21,dec21)).T
    dens2 = knn.fit(X2).eval(Xgrid2).reshape(Ny,Nx)
    Nimage2[i,:,:]=dens2
```

```python
Nimagemax, Nimagemin = max(np.max(Nimage1),np.max(Nimage2)), min(np.min(Nimage1),np.min(Nimage2))
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax

for i in range(Nt):
    im1 = ax[0][i].imshow(Nimage1[i,:,:],origin='lower',vmin=Nimagemin,vmax=Nimagemax)
    im2 = ax[1][i].imshow(Nimage2[i,:,:],origin='lower',vmin=Nimagemin,vmax=Nimagemax)
    ax[0][i].set_title('Bin '+str(i))
    
ax[0][0].set_ylabel('A')
ax[1][0].set_ylabel('B')
    
cbar_ax2 = f.add_axes([0.95, 0.15, 0.02, 0.7])
f.colorbar(im2,cax=cbar_ax2)
```

Smoothing photon images with nearest neighbor (N Neighbor=100).

```python
Ndiff1 = np.zeros_like(Nimage1)
Ndiff2 = np.zeros_like(Nimage2)
for i in range(Nt):
    Ndiff1[i,:,:] = Nimage1[i,:,:]-Nimage1[0,:,:]
    Ndiff2[i,:,:] = Nimage2[i,:,:]-Nimage2[0,:,:]
```

```python
Ndiffmax, Ndiffmin = max(np.max(Ndiff1),np.max(Ndiff2)), min(np.min(Ndiff1),np.min(Ndiff2))
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax

for i in range(Nt):
    im1 = ax[0][i].imshow(Ndiff1[i,:,:],origin='lower',vmin=Ndiffmin,vmax=Ndiffmax)
    im2 = ax[1][i].imshow(Ndiff2[i,:,:],origin='lower',vmin=Ndiffmin,vmax=Ndiffmax)
    ax[0][i].set_title('Bin '+str(i))
    
ax[0][0].set_ylabel('A')
ax[1][0].set_ylabel('B')
    
cbar_ax2 = f.add_axes([0.95, 0.15, 0.02, 0.7])
f.colorbar(im2,cax=cbar_ax2)
```

Residual after subtracting the image of Bin 0.

```python
ratio1 = np.zeros(3)
ratio2 = np.zeros(3)
for i in range(1,Nt):
    ratio1[i-1] = np.max(np.abs(Ndiff1[i,:,:]))/np.max(np.abs(Nimage1[i,:,:]))
    ratio2[i-1] = np.max(np.abs(Ndiff2[i,:,:]))/np.max(np.abs(Nimage2[i,:,:]))
```

```python
a,=plt.plot(range(1,4),ratio1,'r+')
b,=plt.plot(range(1,4),ratio2,'gx')
plt.xlim([0.5,3.5]);plt.xlabel('Bin');plt.ylabel('Residual / Full amplitude')
plt.legend((a,b),('A','B'),loc=2)
plt.yscale('log')
```

```python
print ratio1, ratio2
```

Kernel Density

```python
kernel = 'gaussian' # exponential, tophat

h1=(xmax1-xmin1)/20
kde1 = KernelDensity(h1, kernel=kernel)

h2=(xmax2-xmin2)/20
kde2 = KernelDensity(h2, kernel=kernel)
```

```python
Nimage1 = np.zeros((Nt,Nx,Ny))
Nimage2 = np.zeros((Nt,Nx,Ny))
```

```python
for i in range(Nt):
    ra11,dec11 = ra1[t_table1[i,:]], dec1[t_table1[i,:]]
    X1 = np.vstack((ra11,dec11)).T
    log_dens1 = kde1.fit(X1).score_samples(Xgrid1)
    dens1 = X1.shape[0] * np.exp(log_dens1).reshape((Ny, Nx))
    Nimage1[i,:,:]=dens1
    
    ra21,dec21 = ra2[t_table2[i,:]], dec2[t_table2[i,:]]
    X2 = np.vstack((ra21,dec21)).T
    log_dens2 = kde2.fit(X2).score_samples(Xgrid2)
    dens2 = X2.shape[0] * np.exp(log_dens2).reshape((Ny, Nx))
    Nimage2[i,:,:]=dens2
```

```python
Nimagemax, Nimagemin = max(np.max(Nimage1),np.max(Nimage2)), min(np.min(Nimage1),np.min(Nimage2))
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax

for i in range(Nt):
    im1=ax[0][i].imshow(Nimage1[i,:,:],origin='lower',vmin=Nimagemin,vmax=Nimagemax)
    im2=ax[1][i].imshow(Nimage2[i,:,:],origin='lower',vmin=Nimagemin,vmax=Nimagemax)
    ax[0][i].set_title('Bin '+str(i))
    
ax[0][0].set_ylabel('A')
ax[1][0].set_ylabel('B')
    
cbar_ax2 = f.add_axes([0.95, 0.15, 0.02, 0.7])
f.colorbar(im2,cax=cbar_ax2)
```

Smoothing photon images with kernel density estimation (Gaussian kernel with bandwidth=1/20 of the image width).

```python
Ndiff1 = np.zeros_like(Nimage1)
Ndiff2 = np.zeros_like(Nimage2)
for i in range(Nt):
    Ndiff1[i,:,:] = Nimage1[i,:,:]-Nimage1[0,:,:]
    Ndiff2[i,:,:] = Nimage2[i,:,:]-Nimage2[0,:,:]
```

```python
Ndiffmax, Ndiffmin = max(np.max(Ndiff1),np.max(Ndiff2)), min(np.min(Ndiff1),np.min(Ndiff2))
```

```python
f,ax = plt.subplots(2,4,figsize=(19,9))
((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = ax

for i in range(Nt):
    im1=ax[0][i].imshow(Ndiff1[i,:,:],origin='lower',vmin=Ndiffmin,vmax=Ndiffmax)
    im2=ax[1][i].imshow(Ndiff2[i,:,:],origin='lower',vmin=Ndiffmin,vmax=Ndiffmax)
    
cbar_ax1 = f.add_axes([0.95, 0.55, 0.02, 0.35])
cbar_ax2 = f.add_axes([0.95, 0.13, 0.02, 0.35])
f.colorbar(im1,cax=cbar_ax1)
f.colorbar(im2,cax=cbar_ax2) 
```

Residual after subtracting the image of Bin 0.

```python
ratio1 = np.zeros(3)
ratio2 = np.zeros(3)
for i in range(1,Nt):
    ratio1[i-1] = np.max(np.abs(Ndiff1[i,:,:]))/np.max(np.abs(Nimage1[i,:,:]))
    ratio2[i-1] = np.max(np.abs(Ndiff2[i,:,:]))/np.max(np.abs(Nimage2[i,:,:]))
```

```python
a,=plt.plot(range(1,4),ratio1,'r+')
b,=plt.plot(range(1,4),ratio2,'gx')
plt.xlim([0.5,3.5]);plt.xlabel('Bin');plt.ylabel('Residual / Full amplitude')
plt.legend((a,b),('A','B'),loc=2)
plt.yscale('log')
```

```python
print ratio1, ratio2
```

Our experiments show that all three methods (2D histogram, nearest neighbor, kernel density estimation) smooth photon images fairly well. Ratio plots show that differences between image A and B ($H_1$ vs $H_0$) appeear to be much greater than differences within image B ($H_0$). These are positive signs that all three methods provide satisfying performance in indentifying time-varying sources. However, future work is needed to confirm statistical significance of our findings.
